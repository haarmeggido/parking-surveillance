```python
import json
import os
import time
```

```python
json_path =  './results/results.json'


####
    # {
    #     "id": 0,
    #     "name": "2024_05_07_13_41_0.png",
    #     "year": "2024",
    #     "month": "05",
    #     "day": "07",
    #     "hour": "13",
    #     "minute": "41",
    #     "second": "0",
    #     "left": 0,
    #     "right": 1,
    #     "center": 1
    # },
#####

# read the json file
verbose = False
with open(json_path, 'r') as f:
    results = json.load(f)
    i = 0

    # create a 5 element queue that will store the last 5 results, and will iterate over the results with the new result replacing the oldest one
    queue = [0, 0, 0, 0, 0]
    while i < len(results):
        arr_left = [0, 0, 0, 0, 0]
        arr_right = [0, 0, 0, 0, 0]
        arr_center = [0, 0, 0, 0, 0]

        queue.pop(0)
        queue.append(results[i])
        if verbose:
          for j in range(5):
              print(queue[j])
          print('\n')
        if i > 4:
            # fill the arrays of parking spots with the values of the last 5 results
            for j in range(5):
                arr_left[j] = queue[j]['left']
                arr_right[j] = queue[j]['right']
                arr_center[j] = queue[j]['center']

            print('left: ', arr_left)
            print('right: ', arr_right)
            print('center: ', arr_center)
            
        time.sleep(1)
        i += 1

        if i > 10:
            break
```
