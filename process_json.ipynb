```python
import json
import os
import time
```

```python
json_path =  './results/results.json'


####
    # {
    #     "id": 0,
    #     "name": "2024_05_07_13_41_0.png",
    #     "year": "2024",
    #     "month": "05",
    #     "day": "07",
    #     "hour": "13",
    #     "minute": "41",
    #     "second": "0",
    #     "left": 0,
    #     "right": 1,
    #     "center": 1
    # },
#####

# read the json file
verbose = False
processed_results = []
with open(json_path, 'r') as f:
    results = json.load(f)
    i = 0

    # create a 5 element queue that will store the last 5 results, and will iterate over the results with the new result replacing the oldest one
    queue = [0, 0, 0, 0, 0]
    while i < len(results):
        arr_left = [0, 0, 0, 0, 0]
        arr_right = [0, 0, 0, 0, 0]
        arr_center = [0, 0, 0, 0, 0]

        queue.pop(0)
        queue.append(results[i])
        if verbose:
          for j in range(5):
              print(queue[j])
          print('\n')
        if i > 4:
            # fill the arrays of parking spots with the values of the last 5 results
            for j in range(5):
                arr_left[j] = queue[j]['left']
                arr_right[j] = queue[j]['right']
                arr_center[j] = queue[j]['center']

            # print('left: ', arr_left)
            # print('right: ', arr_right)
            # print('center: ', arr_center)

            # for all three arrays:
            # if arr[0], arr[1], arr[3], arr[4] are all 0, but arr[2] is 1, then it's model error, change arr[2] value
            # if arr[0], arr[1], arr[3], arr[4] are all 1, but arr[2] is 0, then it's model error, change arr[3] value
      
            # left
            if arr_left[0] == 0 and arr_left[1] == 0 and arr_left[3] == 0 and arr_left[4] == 0 and arr_left[2] == 1:
                queue[2]['left'] = 0
                print(f'{i} error in left, changing value to 0')
            if arr_left[0] == 1 and arr_left[1] == 1 and arr_left[3] == 1 and arr_left[4] == 1 and arr_left[2] == 0:
                queue[2]['left'] = 1
                print(f'{i} error in left, changing value to 1')

            # right
            if arr_right[0] == 0 and arr_right[1] == 0 and arr_right[3] == 0 and arr_right[4] == 0 and arr_right[2] == 1:
                queue[2]['right'] = 0
                print(f'{i} error in right, changing value to 0')
            if arr_right[0] == 1 and arr_right[1] == 1 and arr_right[3] == 1 and arr_right[4] == 1 and arr_right[2] == 0:
                queue[2]['right'] = 1
                print(f'{i} error in right, changing value to 1')

            # center
            if arr_center[0] == 0 and arr_center[1] == 0 and arr_center[3] == 0 and arr_center[4] == 0 and arr_center[2] == 1:
                queue[2]['center'] = 0
                print(f'{i} error in center, changing value to 0')
            if arr_center[0] == 1 and arr_center[1] == 1 and arr_center[3] == 1 and arr_center[4] == 1 and arr_center[2] == 0:
                queue[2]['center'] = 1
                print(f'{i} error in center, changing value to 1')

        if i > 3:
          processed_results.append(queue[0])
                
            
        i += 1
        
    print('done')

# write the processed results to a new json file
with open('./results/processed_results.json', 'w') as f:
    json.dump(processed_results, f, indent=4)
```

```python
json_path_processed =  './results/processed_results.json'
events = []
with open(json_path_processed, 'r') as f:
    results = json.load(f)
    '''
    there are three parking spots, left, right, and center, represented by 0 and 1 if they are empty or occupied
    track the changes in the parking spots and create events for each change
    actions = ["Take, "Free"]
    events are of the form:
    {
      "timestamp": 2024-05-07 13:41:00,
      "parking_spot": "1",
      "action": "Take"
    }
    '''

    i = 0
    while i < len(results):
        # first result
        if i == 0:
            # take 'Take' events for the first result
            if results[i]['left'] == 1:
                event = {
                    "timestamp": f"{results[i]['year']}-{results[i]['month']}-{results[i]['day']} {results[i]['hour']}:{results[i]['minute']}:{int(results[i]['second']):02d}",
                    "parking_spot": "1",
                    "action": "Take"
                }
                events.append(event)
            if results[i]['right'] == 1:
                event = {
                    "timestamp": f"{results[i]['year']}-{results[i]['month']}-{results[i]['day']} {results[i]['hour']}:{results[i]['minute']}:{int(results[i]['second']):02d}",
                    "parking_spot": "2",
                    "action": "Take"
                }
                events.append(event)
            if results[i]['center'] == 1:
                event = {
                    "timestamp": f"{results[i]['year']}-{results[i]['month']}-{results[i]['day']} {results[i]['hour']}:{results[i]['minute']}:{int(results[i]['second']):02d}",
                    "parking_spot": "3",
                    "action": "Take"
                }
                events.append(event)
        
        # consecutive results
        if i > 0:
            if results[i]['left'] != results[i-1]['left']:
                event = {
                    "timestamp": f"{results[i]['year']}-{results[i]['month']}-{results[i]['day']} {results[i]['hour']}:{results[i]['minute']}:{int(results[i]['second']):02d}",
                    "parking_spot": "1",
                    "action": "Take" if results[i]['left'] == 1 else "Free"
                }
                events.append(event)
            if results[i]['right'] != results[i-1]['right']:
                event = {
                    "timestamp": f"{results[i]['year']}-{results[i]['month']}-{results[i]['day']} {results[i]['hour']}:{results[i]['minute']}:{int(results[i]['second']):02d}",
                    "parking_spot": "3",
                    "action": "Take" if results[i]['right'] == 1 else "Free"
                }
                events.append(event)
            if results[i]['center'] != results[i-1]['center']:
                event = {
                    "timestamp": f"{results[i]['year']}-{results[i]['month']}-{results[i]['day']} {results[i]['hour']}:{results[i]['minute']}:{int(results[i]['second']):02d}",
                    "parking_spot": "2",
                    "action": "Take" if results[i]['center'] == 1 else "Free"
                }
                events.append(event)
        i += 1

    print('done')

# write the events to a new json file
with open('./results/events.json', 'w') as f:
    json.dump(events, f, indent=4)
```

```python
json_path_events =  './results/events.json'

# delete events 'free' and 'take' that are too close in time (less than 30 seconds)

with open(json_path_events, 'r') as f:
    events = json.load(f)
    i = 0
    while i < len(events):
        if i > 0:
            # check if the current event is 'free' and the previous event is 'take' for the same parking spot
            if events[i]['action'] == 'Free' and events[i-1]['action'] == 'Take' and events[i]['parking_spot'] == events[i-1]['parking_spot']:
                # get the timestamps of the events
                ts1 = time.mktime(time.strptime(events[i-1]['timestamp'], '%Y-%m-%d %H:%M:%S'))
                ts2 = time.mktime(time.strptime(events[i]['timestamp'], '%Y-%m-%d %H:%M:%S'))
                # check if the difference between the timestamps is less than 30 seconds
                if ts2 - ts1 < 30:
                    # delete the events
                    del events[i-1]
                    del events[i-1]
                    i -= 1
            # check if the current event is 'take' and the previous event is 'free' for the same parking spot
            if events[i]['action'] == 'Take' and events[i-1]['action'] == 'Free' and events[i]['parking_spot'] == events[i-1]['parking_spot']:
                # get the timestamps of the events
                ts1 = time.mktime(time.strptime(events[i-1]['timestamp'], '%Y-%m-%d %H:%M:%S'))
                ts2 = time.mktime(time.strptime(events[i]['timestamp'], '%Y-%m-%d %H:%M:%S'))
                # check if the difference between the timestamps is less than 30 seconds
                if ts2 - ts1 < 30:
                    # delete the events
                    del events[i-1]
                    del events[i-1]
                    i -= 1
        i += 1

# write the events to a new json file
with open('./results/events_processed.json', 'w') as f:
    json.dump(events, f, indent=4)
```
