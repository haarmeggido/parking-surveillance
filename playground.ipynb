```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import numpy as np

model = YOLO("yolov8s.pt")  # Load the pre-trained model
```

```python
results = model("car_parking.png")  # Make predictions on the input image
results = results[0]
```

```python
# Display the image with the predicted bounding boxes
# results.show()

# Get the bounding boxes and the corresponding class labels
boxes = results.boxes

for box in boxes:
    print(box.xyxy)

# Get the class labels
labels = results.names

```

```python
Img = plt.imread("car_parking.png")
```

```python
# Display the image with the predicted bounding boxes
plt.imshow(Img)

for box in boxes:
    tensor = box.xyxy.flatten()
    x1 = tensor[0]
    y1 = tensor[1]
    x2 = tensor[2]
    y2 = tensor[3]
    plt.plot([x1, x2, x2, x1, x1], [y1, y1, y2, y2, y1], 'r', lw=2)

plt.show()
```
